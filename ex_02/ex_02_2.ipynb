```python
import cv2
import numpy as np
import time
import os
import matplotlib.pylab as plt
%matplotlib inline
```

```python
img_save_path = 'calisma_2_2/'

#create folder ofnot existing
try:
    os.stat(img_save_path)
except:
    os.mkdir(img_save_path)       


vcap = cv2.VideoCapture(0)
```

```python
for i in range(5):
    for j in range(5):
        ret,image = vcap.read()
        if ret:        
            plt.axis('off')
            #plt.figure(figsize=(20,44))
            plt.imshow(image)
            plt.show()
            path = "{}sn{}.{}.jpg".format(img_save_path,i,j*20)
            cv2.imwrite(path,image)
            time.sleep(1/5)
        else:
            print('Image Not readed ')
vcap.release()
```
